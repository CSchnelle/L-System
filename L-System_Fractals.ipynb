The Lindenmeyer system is a parallel re-writing system. A rewriting system is made up of a wide range of methods that use substitution to replace subterms of a formula with other terms. In other words, you start with a set of objects and a rewriting sytem will give instructions 
for relations and how to transform those objects. An example of a rewriting system would be Conjunctive Normal Form (CNF), in formal logic, can be used as a rewriting system. The L-System is also a type of formal grammar, consisting of an alphabet that is used to make strings,production rules 
(i.e.rewrite rules that specify symbol substitution recursively performed to generate new symbols.), an initial axiom (i.e. starting point) and a method to generate strings into geometric structures. The L-System or Lindenmayer system, developmed by Aristid Lindenmayer, a botanist and biologist, originally used this L-System to describe 
the behavior of plant development but it can also be used more generally to create fractals that are self-similar due to its recursive nature and can describe a variety of morphologies. In a programming language an l-system can be represented as a tuple, where the first element is the axiom and the second is a dictionary mapping the original symbols to their substitution strings. (i.e. l_system = ("A", {"A":"AB"})). An example of the structure is shown below.

Here is an example of a simple L-System following the initial axiom and rules to substitute and create a new string of symbols.

```python
def l_system(n, axiom="A", rules={"A": "AB", "B": "A"}):
    state = axiom
    for _ in range(n):
        state = "".join([rules.get(ch, ch) for ch in state])
    return state

# Run it for 5 iterations
print(l_system(5))
```

Using matplot and numpy we can create a tree structure using the L-System

```python
import matplotlib.pyplot as plt
import numpy as np
# Tree Structure
# Define the L-system rules
rules = {"F" : "FF+[+F-F-F]-[-F+F+F]"}

# Initial condition (axiom)
axiom = "F"

# Number of iterations
n = 4

# Angle in degrees
angle = 22.5


# Start position
x, y = 0, 0

# Start direction
theta = 90

# Positions
lines = []

# Stack for saved states during branching
stack = []

# Apply the rules for n iterations
state = axiom
for _ in range(n):
    state = "".join([rules.get(ch, ch) for ch in state])

# Draw the fractal
for command in state:
    if command == "F":
        # Calculate new position
        new_x = x + np.cos(np.radians(theta))
        new_y = y + np.sin(np.radians(theta))
        # Draw line to new position
        lines.append(((x, y), (new_x, new_y)))
        x, y = new_x, new_y
    elif command == "[":
        # Save state
        stack.append((x, y, theta))
    elif command == "]":
        # Load saved state
        x, y, theta = stack.pop()
    elif command == "+":
        # Turn right
        theta -= angle
    elif command == "-":
        # Turn left
        theta += angle

# Display the fractal
fig, ax = plt.subplots()
for ((x1, y1), (x2, y2)) in lines:
    ax.plot([x1, x2], [y1, y2], 'k')

plt.axis('equal')
plt.show()
```
